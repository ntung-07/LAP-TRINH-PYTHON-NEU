```python
# Bài 1
n1 = int(input("Nhap so"))
n2 = int(input("Nhap so"))

def check_gcd(n):
    if n<2:
        return False
    else:
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                return False
    return True

print(check_gcd(n1))
print(check_gcd(n2))
```

```python
# Bài 3
lst = eval(input("List:")) # eval() nó exe input luôn, chuyển str thành list thật
lst = list(map(str, lst))
print(lst)
converted = "".join(lst)
print("Converted to int: ", converted)
```

```python
# Bài 4
s = input("Nhap 1 chuoi")

if len(s) >= 3:
    if not s.endswith("ing"):
        s = s + "ing"
        print(s)
    else:
        s = s + "ly"
        print(s)
else:
    print(s)
```

```python
# Bài 5:
n = int(input("Nhập vào 1 số nguyên"))

tong_uoc = 0
for i in range(1, n):
    if n % i == 0:
        tong_uoc += i

if tong_uoc > n:
    print(True)
else:
    print(False)
```

```python
# Bài 6:
n = input("Nhap so ve")
if len(n) % 2 != 0:
    print("Hay nhap so luong chu so chan")

half = len(n) // 2

first_half = n[:half]
second_half = n[half:]

sum_first = sum(int(c) for c in first_half)
sum_second = sum(int(c) for c in second_half)

if sum_first == sum_second:
    print("Đó là số may mắn!")
else:
    print("Đó là số không may mắn")
```

```python
# Bài 7
s1 = input("Nhập vào chuỗi thứ nhất")
s2 = input("Nhập vào chuỗi thứ hai")

count = 0
for c in s1:
    if c in s2:
        count += 1
        s2 = s2.replace(c, "", 1)

print(count)
```
