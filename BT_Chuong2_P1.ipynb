```python
# Bài 1
s = input("Nhap vao 1 chuoi")
n = int(input("Nhap vao 1 so nguyen"))

targ = n-1
res = []

for i in range(len(s)):
    if i <= targ:
        res.append(s[i].lower())
    else:
        res.append(s[i])
print("".join(res))
```

```python
# Bài 2
s = input("Nhập vào 1 chuỗi")

res = []

for c in s:
    if c == " ":
        continue
    else:
        res.append(c)

print("".join(res))
```

```python
# Bài 3
s = input("Nhập vào 1 chuỗi:")
d = input("Nhập ký tự muốn xóa")

res = []

for c in s:
    if c == d:
        continue
    else:
        res.append(c)

print("".join(res))
```

```python
# Bài 5: Check palindrome

def check_palindrome(s):
    left, right = 0, len(s)-1
    while left < right:
        if s[left] == s[right]:
            left += 1
            right -= 1
        else:
            return False
    return True

s = input("Nhập vào 1 chuỗi")
print(check_palindrome(s))
```
