```python
min_index = df_cleaned["Founded"].idxmin()
min = df_cleaned.loc[min_index, "Founded"]
min
```

```python
top5 = df_cleaned.groupby("State")["MarketCap"].sum().nlargest(5)
state_sum = df_cleaned.groupby("State")["MarketCap"].sum()
others = state_sum.sum() - top5.sum()

labels = list(top5.index) + ["others"]
labels
```
