```python
# Câu 1
class Animal:
    def __init__(self, name, genus, species):
        self.name = name
        self.genus = genus
        self.species = species

    def displayInfo(self):
        print("=====Displaying Animal detail:=====\n\n")
        print(f"\nName: {self.name}")
        print(f"\nGenus: {self.genus}")
        print(f"\nSpecies: {self.species}")
        print("=================end===================")

class Dog(Animal):
    def __init__(self, name, genus, species, color):
        super().__init__(name, genus, species)
        self.color = color

    # Dog's own function
    def displayDog(self):
        print("=====Displaying Dog (dog's function) detail:=====")
        print(f"Name: {self.name}")
        print(f"Genus: {self.genus}")
        print(f"Species: {self.species}")
        print(f"Color: {self.color}")

    # Animal's function (overridden)
    def displayInfo(self):
        print("\n\n=====Displaying Dog (Animal Override) detail:=====")
        print(f"Name: {self.name}")
        print(f"Genus: {self.genus}")
        print(f"Species: {self.species}")
        print(f"Color: {self.color}")
    
# Execute test cases
doggo = Dog("Milu", "Shepard", "Sausage", "Captain Black")
doggo.displayDog()
doggo.displayInfo()
```

```python
# Câu 2:

# a.
def is_perfect_square(num: int) -> bool:
    if num >= 0:
        square_root = num ** 0.5
        # See if square_root is a whole number
        return (square_root % 1 == 0)
    return False

# b.
def sum_PerfectSquare(start: int, end: int) -> int:
    sum = 0
    for num in range(start, end + 1):
        if is_perfect_square(num):
            sum += num
    return sum

# c. Execute test cases
print(sum_PerfectSquare(0, 10))     # 14
print(sum_PerfectSquare(1, 20))     # 30
print(sum_PerfectSquare(10, 50))    # 126
print(sum_PerfectSquare(25, 100))   # 355
print(sum_PerfectSquare(100, 200))  # 730
```

```python
# Câu 3:

import pandas as pd
import numpy as np

df = pd.read_csv("books.csv")

# Create dummy column Rating as it is not present in the data
df["Rating"] = np.random.randint(1, 11, df.shape[0])
df = df.dropna()
df
```

```python
import matplotlib.pyplot as plt

# Define what's x and what's y
x_axis = df["Publisher"].unique()
y_axis = df.groupby("Publisher")["Rating"].mean()

# a. Biểu đồ đường
plt.figure(figsize=(12, 8))
plt.plot(x_axis, y_axis)
plt.xticks(rotation=90)
plt.show()
```

```python
# Define what's in the pie
count_book = df["Publisher"].value_counts().nlargest(5)

# b.
plt.figure(figsize=(8, 8))
plt.pie(count_book)
```
