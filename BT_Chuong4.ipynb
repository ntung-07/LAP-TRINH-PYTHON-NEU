```python
# Bài 6
inputArr = eval(input("Nhập dãy tọa độ các chướng ngại vật:"))

def minStep(inputArr) -> int:
    for k in range(2, 100):
        status = "OK"
        for n in inputArr:
            if n % k == 0:
                status = "NXT"
                break
            else:
                continue
        if status == "OK":
            return k
    return "Unsolvable"
minStep(inputArr)
```
