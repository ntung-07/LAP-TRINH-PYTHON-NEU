**Bài 1.1: Gán và in giá trị**
<p>Gán giá trị 10 cho biến a, giá trị 20 cho biến b, sau đó in tổng, hiệu, tích và thương.</p>

```python
a = 10
b = 20
print(a+b)
print(a*b)
print(abs(a-b))
print(a/10)
```

**Bài 1.2: Hoán đổi giá trị của hai biến**
<p>Nhập hai số nguyên từ bàn phím, hoán đổi giá trị của chúng và in kết quả</p>

```python
a = 1
b = 2
temp = a
a = b
b = temp
print(a)
print(b)
```

**Bài 1.3: Tính diện tích và chu vi hình tròn**
<p>Nhập bán kính hình tròn, tính và in diện tích và chu vi.</>

```python
import math
pi_value = math.pi

bankinh = int(input("Nhap ban kinh: "))
s = pi_value * (bankinh ** 2)
print("Dien tich: ", s)
p = 2 * pi_value * bankinh
print("Chu vi: ",p)
```

**Bài 1.4: Vào/ra dữ liệu**
<p>Viết chương trình yêu cầu người dùng nhập tên, tuổi và email của họ. Sau đó, in ra một câu chào mừng có đầy đủ thông tin đã nhập, ví dụ: "Chào [Tên], bạn [Tuổi] tuổi và có email [Email]. Đảm bảo sử dụng các kiểu dữ liệu phù hợp cho từng thông tin.</>

```python
ten = input("Nhap ten: ")
tuoi = int(input("Nhap tuoi: "))
email = str(input("Nhap email: "))

print(f"Chao {ten}, ban {tuoi} tuoi, va co email {email}.")
print("Chao {}, ban {} tuoi, va co email {}".format(ten, tuoi, email))
```
