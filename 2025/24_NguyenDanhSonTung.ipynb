```python
# Tạo lớp Country với các thuộc tính
class Country:
    def __init__(self, name, capital, population):
        self.name = name                # tên quốc gia
        self.capital = capital          # thủ đô
        self.population = population    # dân số

    # Tạo phương thức hiển thị thông tin
    def displayInfo(self):
        print("\n=====Đang hiển thị=====")
        print(f"Tên quốc gia: {self.name}")
        print(f"Thủ đô: {self.capital}")
        print(f"Dân số: {self.population}")

# Tạo lớp con DevelopedCountry kế thừa lớp Country
class DevelopedCountry(Country):
    def __init__(self, name, capital, population, gdp):
        super().__init__(name, capital, population)
        self.gdp = gdp  # Có thêm thuộc tính gdp

    # Câu 2:
    # Kế thừa phương thức displayInfo
    def displayInfo(self):
        super().displayInfo()
        print(f"GDP: {self.gdp}") # ghi đè thêm gdp

    # Tạo phương thức hiển thị thông tin riêng
    def displayDevelopedCountry(self):
        print("\n===Hiển thị thông tin quốc gia phát triển===")
        print(f"Tên QG: {self.name}\nThủ đô: {self.capital}\nDân số: {self.population}\nGDP: {self.gdp}")

# Tạo các đối tượng và gọi phương thức để test
country_1 = Country("Vietnam", "Hanoi", 100000000)
country_2 = DevelopedCountry("Japan", "Tokyo", 200000000, 998999)

country_1.displayInfo()
country_2.displayInfo()
country_2.displayDevelopedCountry()
```

```python
# Câu 3: Kiểm tra danh sách tăng dần
def isIncreasing(num_list) -> bool:
    for i in range(len(num_list) - 1):
        if num_list[i] >= num_list[i + 1]:
            return False
    return True

# Câu 4: Dãy con liên tiếp dài nhất
def findLongestIncreasingSubsequence(input_list):
    best = []
    for i in range(len(input_list)):
        for j in range(i+1, len(input_list)+1):
            sub = input_list[i:j]
            if isIncreasing(sub) and len(sub) > len(best):
                best = sub
    return best

# Câu 5: Chạy thử các test cases
print(findLongestIncreasingSubsequence([1, 2, 3, 4, 5, 1, 2, 3]))
print(findLongestIncreasingSubsequence([5, 4, 3, 2, 1]))
print(findLongestIncreasingSubsequence([3, 10, 2, 1, 20]))
print(findLongestIncreasingSubsequence([10, 20, 30, 10, 20, 40]))
print(findLongestIncreasingSubsequence([4, 6, 7, 8, 10, 12, 2]))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Câu 6
# Đọc dữ liệu CSV vào DataFrame
df = pd.read_csv("books.csv")
df
```

```python
# Tính trung bình Rating theo Author
avg_rating = df.groupby("Author")["Rating"].mean()

# Vẽ biểu đồ cột
plt.figure(figsize=(10, 6))
plt.bar(avg_rating.index, avg_rating.values)
plt.xlabel("Author")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Câu 7
# Tổng số lượng sách của từng tác giả
book_count = df["Author"].value_counts()

# Vẽ biểu đồ tròn
plt.figure(figsize=(8, 8))
plt.pie(book_count.values, labels=book_count.index, autopct="%1.1f%%")
plt.tight_layout()
plt.show()
```
