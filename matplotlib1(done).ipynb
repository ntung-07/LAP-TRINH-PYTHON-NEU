Đọc dữ liệu từ dataset <b>Company_sales.csv</b> và lưu vào DataFrame

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("company_sales.csv")
df
```

1. Đọc tổng lợi nhuận (<b>Total profit</b>) của tất cả các tháng (<b>month_number</b>) và hiển thị nó bằng cách sử dụng biểu đồ dạng <b>line</b>.

```python
plt.figure(figsize=(10, 6))
plt.plot(df["month_number"], df["total_profit"], label="Tổng lợi nhuận theo tháng")
plt.title("Tổng lợi nhuận theo tháng")
plt.xlabel("Tháng")
plt.ylabel("Tổng lợi nhuận")
plt.legend()
plt.show()
```

2. Tính tổng số lượng bán được (<b>Total unit</b>) của tất cả các tháng và hiển thị biểu đồ đường với các thuộc tính:
- Kiểu đường (Line Style) nét đứt và màu đường (Line-color) phải là màu đỏ
- Hiển thị chú giải (legend) ở vị trí phía dưới bên phải.
- Nhãn trục X = Month Number
- Nhãn trục Y = Sold units number
- Thêm điểm đánh dấu (marker) vòng tròn.
- Độ rộng đường (Linewidth) là 3

```python
plt.figure(figsize=(10, 6))
plt.plot(
    df["month_number"],
    df["total_units"],
    linestyle="--",
    color="red",
    marker="o",
    linewidth=3,
    label="Total Units"
)
plt.xlabel("Month number")
plt.ylabel("Sold units number")
plt.legend(loc="lower right")
plt.show()
```

3. Đọc tất cả dữ liệu bán sản phẩm và hiển thị nó bằng cách sử dụng một biểu đồ nhiều đường (line). Hiển thị số lượng bán được mỗi tháng cho mỗi sản phẩm.

```python
plt.figure(figsize=(12, 8))
plt.plot(df["month_number"], df["facecream"], label="Face cream", marker="o")
plt.plot(df["month_number"], df["facewash"], label="Face wash", marker="s")
plt.plot(df["month_number"], df["toothpaste"], label="Tooth paste", marker="^")
plt.plot(df["month_number"], df["bathingsoap"], label="Bathing soap", marker="d")
plt.plot(df["month_number"], df["shampoo"], label="Shampoo", marker="*")
plt.plot(df["month_number"], df["moisturizer"], label="Moisturizer", marker="p")
plt.title("Dữ liệu bán tất cả sản phẩm theo tháng")
plt.xlabel("Tháng")
plt.ylabel("Số lượng bán được")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

4. Đọc dữ liệu bán kem đánh răng (toothpaste) của mỗi tháng và hiển thị dữ liệu đó bằng cách sử dụng biểu đồ phân tán. Thêm đường lưới trong biểu đồ, kiểu đường lưới (gridline style) là ‘--'.

```python
plt.figure(figsize=(10, 6))
plt.scatter(df["month_number"], df["toothpaste"], label="toothpaste", s=100, color="r")
plt.title("Dữ liệu bán kem đánh răng theo tháng")
plt.xlabel("Tháng")
plt.ylabel("Số lượng kem đánh răng bán được")
plt.grid(True, alpha=0.7, linestyle="--")
plt.show()
```

5. Đọc dữ liệu bán sản phẩm kem mặt (Face Cream) và sữa rửa mặt (Face wash) và hiển thị dữ liệu đó bằng cách sử dụng biểu đồ thanh (bar).
Biểu đồ thanh sẽ hiển thị số lượng đơn vị bán được mỗi tháng cho mỗi sản phẩm (thanh riêng biệt cho từng sản phẩm trong cùng một biểu đồ).

```python
plt.figure(figsize=(12, 6))
months = df['month_number']
width = 0.35
plt.bar(months - width/2, df["facecream"], width=0.35, label="Face cream", color="skyblue")
plt.bar(months + width/2, df["facewash"], width=0.35, label="Face wash", color="orange")
plt.title("Dữ liệu bán Face cream và Face wash theo tháng")
plt.xlabel("Tháng")
plt.ylabel("Số lượng bán được")
plt.legend()
plt.xticks(df["month_number"])
plt.grid(True, alpha=0.3, axis="y")
plt.show()
```

7. Đọc tổng lợi nhuận của mỗi tháng và hiển thị nó bằng cách sử dụng biểu đồ để xem các phạm vi lợi nhuận phổ biến nhất

```python
plt.figure(figsize=(10, 6))
plt.hist(df["total_profit"], bins="auto", color="skyblue", edgecolor="black", alpha=0.7)
plt.title("Phân bổ tổng lợi nhuận theo tháng")
plt.xlabel("Tổng lợi nhuận")
plt.ylabel("Tần suất")
plt.grid(True, alpha=0.3, axis="y")
plt.show()
```

8. Tính toán tổng dữ liệu bán hàng cho từng sản phẩm và hiển thị dữ liệu đó bằng biểu đồ pie.

```python
product_totals= [
    df["facecream"].sum(),
    df["facewash"].sum(),
    df["toothpaste"].sum(),
    df["bathingsoap"].sum(),
    df["shampoo"].sum(),
    df["moisturizer"].sum()
]

product_names = ["Face cream", "Face wash", "Toothpaste", "Bathing soap", "Shampoo", "Moisturizer"]

plt.figure(figsize=(10, 8))
plt.pie(product_totals, labels=product_names, autopct="%1.1f%%", startangle=90)
plt.title("Tổng dữ liệu bán hàng cho từng sản phẩm")
plt.axis("equal") # Đảm bảo biểu đồ tròn
plt.show()
```

9. Đọc sữa rửa mặt (facewash), xà phòng tắm (bathingsoap) của tất cả các tháng và hiển thị nó bằng Subplot

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6)) # Tạo 1 hàng, 2 cột

ax1.plot(df["month_number"], df["facewash"], marker="o", color="blue")
ax1.set_title("Dữ liệu bán Face wash theo tháng")
ax1.set_xlabel("Tháng")
ax1.set_ylabel("Số lượng bán được")
ax1.grid(True, alpha=0.3)

ax2.plot(df["month_number"], df["bathingsoap"], marker="s", color="green")
ax2.set_title("Dữ liệu bán Bathing soap theo tháng")
ax2.set_xlabel("Tháng")
ax2.set_ylabel("Số lượng bán được")
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

10. Đọc tất cả dữ liệu bán sản phẩm của từng loại và hiển thị dữ liệu đó bằng cách sử dụng biểu đồ ngăn xếp (stack)

```python
plt.figure(figsize=(12, 8))

products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc', '#c2c2f0']

bottom = np.zeros(len(df)) # Khởi tạo đáy cho stack

for i, product in enumerate(products):
    plt.bar(df["month_number"], df[product], bottom=bottom,
            label=product.replace("_", " ").title(), color=colors[i])
    bottom += df[product]

plt.title("Dữ liệu bán tất cả sản phẩm theo tháng")
plt.xlabel("Tháng")
plt.ylabel("Số lượng bán được")
plt.legend()
plt.xticks(df["month_number"])
plt.grid(True, alpha=0.3, axis="y")
plt.show()
```
