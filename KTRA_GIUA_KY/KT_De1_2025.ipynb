````python
import string

# Optional: List out the hexadecimal chars
# Benefit: independence from external libraries
hexdigits = "0123456789abcdef"

def isMACAddress(inputString) -> bool:
    count = 0
    for i in range(len(inputString)):
        if inputString[i] in string.hexdigits:
            count += 1
        
        if inputString[i] == "-":
            if count == 2:
                count = 0
                continue
            else:
                return False
    return True
````

````python
inputString = "00-1B-63-84-45-E6"
print(isMACAddress(inputString))
inputString = "Z1-1B-63-84-45-E6"
print(isMACAddress(inputString))
inputString = "not a MAC-48 address"
print(isMACAddress(inputString))
````

**Definition**

````python
class Book:
    def __init__(self, book_name: str, book_author: str, book_stock: int):
        self.book_name = book_name
        self.book_author = book_author
        self.book_stock = book_stock

    def on_borrow_book(self) -> bool:
        if self.book_stock > 0:
            self.book_stock -= 1
            print(f"200 - Successfully borrowed {self.book_name}. Happy reading!")
            print(f"Left in stock: {self.book_stock}")
            return True
        else:
            print(f"Unsuccessful: No {self.book_name} left in stock.")
            return False

    def on_return_book(self):
        self.book_stock += 1
        print("Successful. Thank you for returning")
        print(f"Updated '{self.book_name} stock: {self.book_stock}'")

    def check_stock(self):
        print(f"Checking '{self.book_name}'...")
        print(f"Stock returned for {self.book_name}: {self.book_stock}")
        if self.book_stock > 0:
            return True
````

````python
class Library:
    def __init__(self):
        # List of books
        self.books = []

        # Advanced: Dictionary of books
        # Benefit: May or may no be easier to manage
        # self.books = {}

    def add_book(self, book_instance: Book):
        self.books.append(book_instance)
        print(f"200 - Successfully added '{book_instance.book_name}' to the library.")

    def remove_book(self, book_name):
        for book in self.books:
            if book.book_name == book_name:
                self.books.remove(book)
                print(f"200 - Removed {book_name} from the library.")
                return
        print(f"Error: '{book_name} not found in the library.'")

    def get_book_info(self, book_name):
        for book in self.books:
            if book.book_name == book_name:
                print(f"Name: {book.book_name}")
                print(f"Author: {book.book_author}")
                print(f"Remaining: {book.book_stock}")
````

````python
class Member:
    def __init__(self, member_id: str, member_name: str):
        self.member_id = member_id
        self.member_name = member_name
        self.books_borrowed = []
    
    def borrow_book(self, library_instance: Library, book_name):
        for book in library_instance.books:
            if book.book_name == book_name:
                ok = book.on_borrow_book()
                if ok:
                    self.books_borrowed.append(book)
                else:
                    break

    def return_book(self, book_name):
        for book in self.books_borrowed:
            if book.book_name == book_name:
                book.on_return_book()
                self.books_borrowed.remove(book)
````

**Execution**

````python
library = Library()
library.add_book(Book("Python Programming", "John Smith", 5))
library.add_book(Book("Data Science Essentials", "Alice Johnson", 3))
````

````python
mem1 = Member("M1", "Rowan Atkinson")
mem2 = Member("M2", "Tô Lâm")
````

````python
# Thành viên 1 mượn một cuốn sách từ thư viện
print("Thành viên 1 mượn một cuốn sách từ thư viện\n")
mem1.borrow_book(library, "Python Programming")
library.get_book_info("Python Programming")

# Thành viên 2 mượn một cuốn sách từ thư viện
print("\n==============================")
print("Thành viên 2 mượn một cuốn sách từ thư viện\n")
mem2.borrow_book(library, "Python Programming")
library.get_book_info("Python Programming")

# Thành viên 1 trả cuốn sách đã mượn
print("\n==============================")
print("Thành viên 1 trả cuốn sách đã mượn\n")
mem1.return_book("Python Programming")
library.get_book_info("Python Programming")

# Tìm và hiển thị thông tin về một cuốn sách cụ thể trong thư viện
print("\n==============================")
library.get_book_info("Data Science Essentials")
print("\n")
library.get_book_info("Python Programming")
````

3.3. Tính tổng số công ty theo từng Sector.

3.4. Liệt kê 5 bang (State) có số lượng công ty nhiều nhất

3.5. Công ty có MarketCap lớn nhất và nhỏ nhất là gì? In ra Company, MarketCap, State, Sector

````python
# Index of the company that holds largest MarketCap
max_marketcap_index = df["MarketCap"].idxmax()

# Index of the company that holds smallest MarketCap
min_marketcap_index = df["MarketCap"].idxmin()

# Retrieve information of the companies via indices
max_company = df.loc[max_marketcap_index] # object
min_company = df.loc[min_marketcap_index] # object

resulting_df = pd.DataFrame([max_company, min_company])
# Explanation: [max_company, min_company] is a list of 2 objects
# this converts the list into a DataFrame

output_columns = ["Company", "MarketCap", "State", "Sector"]
final_result = resulting_df[output_columns].reset_index(drop=True)

print("Thông tin công ty có MarketCap lớn nhất và nhỏ nhất:")
final_result
````

3.6. Tính giá trị trung bình và độ lệch chuẩn (mean và std) của Revenue theo từng Sector

````python
revenue_stats = df.groupby("Sector")["Revenue"].agg(['mean', 'std']).reset_index()

# Optional: Rename columns:
# revenue_stats.columns = ["Sector", "Mean", "Standard Deviation"]

display(revenue_stats)
````

3.7. Tìm Sector có Revenue trung bình cao nhất

````python
# Group by Sector and calculate mean Revenue
sector_max_revenue = df.groupby("Sector")["Revenue"].mean()

# Find the highest via index
highest_revenue_sector = sector_max_revenue.idxmax()
highest_revenue_value = sector_max_revenue.max()

# display(sector_max_revenue)
print(f"Sector có Revenue trung bình cao nhất là: {highest_revenue_sector}")
print(f"với giá trị Revenue trung bình: {highest_revenue_value}")
````

3.8. Tìm công ty thành lập gần đây nhất và công ty lâu đời nhất, in ra tên công ty và năm thành lập

````python
# df.loc[row, column]
newest_company = df.loc[df["Founded"].idxmax(), ["Company", "Founded"]]
oldest_company = df.loc[df["Founded"].idxmin(), ["Company", "Founded"]]

print(f"Công ty lâu đời nhất:\n\n{oldest_company}")
print(f"\n\nCông ty mới thành lập:\n\n{newest_company}")
````

3.9. Tìm State có nhiều công ty thuộc lĩnh vực AI/ML nhất, và số công ty AI/ML của bang đó.

````python
AI_ML_df = df[df["Sector"] == "AI/ML"]
AI_ML_statecount = AI_ML_df.groupby("State").size()

# Search by index
top_aiml_state = AI_ML_statecount.idxmax()
comp_count = AI_ML_statecount.max()

print(f"State có nhiều công ty AI/ML nhất: {top_aiml_state}")
print(f"Số lượng công ty: {comp_count}")
````

- Recap: Có thể nhận thấy 1 pattern tư duy khi làm các bài này đó là, thu hẹp DataFrame rồi search trên nó sử dụng idxmin, idxmax

3.11. Vẽ biểu đồ cột thể hiện 10 công ty có MarketCap lớn nhất. Dãn nhãn cột bằng Stock (hoặc Company nếu chưa IPO). Thêm nhãn trục, tiêu đề, và tùy chỉnh màu sắc

````python
# using matplotlib
# x: stock (=company nếu chưa IPO)
# y: MarketCap

# Step 1: Narrow down the scope
top_10_marketcap = df.sort_values(by="MarketCap", ascending=False).nlargest(10, columns="MarketCap").copy()

# Step 2: (Optional) create a new column named Label, define logic for its content
top_10_marketcap["Label"] = top_10_marketcap.apply(
    lambda row: row["Stock"] if pd.notna(row["Stock"]) else row["Company"],
    axis=1
)

# Visualize resulting dataframe
top_10_marketcap
````

````python
# using matplotlib
# x: stock (=company nếu chưa IPO)
# y: MarketCap

# x
label = top_10_marketcap["Label"]
# y
market_caps = top_10_marketcap["MarketCap"]

plt.figure(figsize=(12, 6))

plt.bar(label,market_caps)
plt.title("10 Công ty có Market Cap lớn nhất (sử dụng matplotlib)")
plt.xlabel("Công ty")
plt.ylabel("Market Cap")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
````

Có thể thấy, do giá trị của từng cột xấp xỉ nhau nên sự chênh lệch không được thể hiện rõ ràng. Có thể làm thêm các bước sau đây:

````python
# Configure y-axis's zoom-in range: from 1200 - 1550
Y_MIN = 1200
Y_MAX = 1550

plt.figure(figsize=(12, 7))
bars = plt.bar(label, market_caps, color="#33AFFF")

# Apply the set limit configurations
plt.ylim(Y_MIN, Y_MAX)

for bar in bars:
    y_val = bar.get_height() # y value = bar's height
    plt.text(bar.get_x() + bar.get_width()/2, y_val + 5, f'{y_val:.2f}', ha="center", va="bottom", fontsize=9)

plt.title("10 Công ty có Market Cap lớn nhất (sử dụng matplotlib)")
plt.xlabel("Công ty")
plt.ylabel("Market Cap")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
````
