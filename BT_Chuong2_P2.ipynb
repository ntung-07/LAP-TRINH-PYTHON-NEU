```python
# Bài 1
s = input("Nhập vào 1 chuỗi các số phân tách bằng dấu phẩy:")

res_list = s.split(",")
res_tuple = tuple(res_list)

print("List: ", res_list)
print("Tuple: ", res_tuple)
```

```python
# Bài 2
s = input("Nhập tên file")

s_list = s.split(".")
print(s_list)

extension = s_list[-1]
print("\n", extension)
```

```python
# Bài 3
n = input("Nhập vào 1 số nguyên:")

nn = int(n*2)
nnn = int(n*3)
n = int(n)

print(n + nn + nnn)
```

```python
# Bài 4
a = int(input("Nhập a"))
b = int(input("Nhập b"))

tong = a + b
hieu = abs(a-b)
tich = a*b
thuong = a/b

res = [a, b, tong, hieu, tich, thuong]

print(res)
```
